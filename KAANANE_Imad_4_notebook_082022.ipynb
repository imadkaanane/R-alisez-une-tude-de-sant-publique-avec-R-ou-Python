## 2.1 - La proportion de personnes en état de sous-nutrition 

```python
# Filtrer les données pour l'année 2017
population_2017 = donnees_population[donnees_population['Année'] == 2017].copy()

# Afficher le DataFrame résultant pour l'année 2017
population_2017
```

```python
# Calculer la somme des populations des pays pour obtenir la population mondiale en 2017
population_totale_2017 = sum(population_2017["Nombre d'habitants"]) * 1e3

population_totale_2017 = int(population_totale_2017)

# Afficher le message avec la population mondiale en 2017
print(f"En 2017, la population mondiale était estimée à environ {population_totale_2017} personnes.")
```

```python
# Filtrage des données pour l'année 2017
donnees_sous_nutri_2017 = donnees_sous_nutrition[donnees_sous_nutrition['Année'] == 2017].copy()
# Affichage du DataFrame résultant
donnees_sous_nutri_2017
```

```python
# Calcul de la somme totale des personnes en sous-nutrition pour l'année 2017, en les ramenant à l'unité (millions)
population_sous_nutri_2017 = sum(donnees_sous_nutri_2017["Population Sous Alimentés"]) * 1e6

# Conversion du résultat en entier pour l'affichage final
population_sous_nutri_2017 = int(population_sous_nutri_2017)

# Affichage du nombre total de personnes en sous-alimentation en 2017
print(f"En 2017, le nombre total de personnes en situation de sous-alimentation était estimé à environ {population_sous_nutri_2017}.")
```

```python
# Calcul de la proportion des personnes en sous-nutrition par rapport à la population totale en 2017
proportion_sous_nutri_2017 = population_sous_nutri_2017 / population_totale_2017 * 100

# Arrondir le résultat à 2 chiffres après la virgule
proportion_sous_nutri_2017 = np.round(proportion_sous_nutri_2017, 2)

# Affichage du résultat
print(f"En 2017, environ {proportion_sous_nutri_2017}% de la population mondiale était touchée par la sous-nutrition.")
```

```python
# Données pour le graphique
labels = ['Population en sous-nutrition', 'Population alimentée']
sizes = [proportion_sous_nutri_2017, 100 - proportion_sous_nutri_2017]

# Création de la figure
fig = px.pie(values=sizes, names=labels, title='Population en sous-nutrition en 2017',
             labels={'names': 'Titre', 'values': 'Pourcentage'},
             hole=0.5, color_discrete_sequence=px.colors.qualitative.Pastel)

# Mise en forme de la figure
fig.update_traces(textinfo='percent+label', pull=[0.2, 0])
fig.update_layout(title={'text': 'Population en sous-nutrition en 2017', 'font': {'size': 18}},
                  template='plotly_dark')
# Afficher l'image dynamique
fig.show()
# Redimensionner l'image 
fig.update_layout(width=1240)
# Enregistrer en tant qu'image
fig.write_image("figure/1.png")
# Afficher l'image sur le depot Github
display(Image(filename="figure/1.png"))
```

## 2.2 - Le nombre théorique de personnes qui pourraient être nourries

```python
# Fusion des données alimentaires avec la population pour l'année 2017
dispo_alim_population = donnees_disponibilite_alimentaire.merge(population_2017, on='Zone')

# On multiplie les valeurs de la colonne "Nombre d'habitants" par 1000 pour les avoir à l'unité
dispo_alim_population["Nombre d'habitants"] = dispo_alim_population["Nombre d'habitants"] * 1000

# Afficher le DataFrame résultant avec les données alimentaires et la population pour 2017
print("DataFrame résultant de la fusion des données alimentaires avec la population pour 2017:")
dispo_alim_population
```

```python
# Calculer la disponibilité alimentaire totale en kilocalories pour chaque pays pour l'année
dispo_alim_population_annee = dispo_alim_population.apply(
    lambda row: row['Disponibilité alimentaire (Kcal/personne/jour)']
    * row["Nombre d'habitants"]
    * 365,
    axis=1
)

# Calculer la somme totale des disponibilités alimentaires mondiales pour une année
dispo_alim_population_annee_total = dispo_alim_population_annee.sum()

 # Supposition : Une personne a besoin d'environ 2500 Kcal par jour en moyenne
besoin_calorique_journalier = 2500

# Calculer le nombre total de personnes pouvant être nourries avec les disponibilités alimentaires mondiales
nombre_population = np.round(dispo_alim_population_annee_total / (besoin_calorique_journalier * 365), 2)

# Calculer le pourcentage de la population mondiale que représente ce nombre
pourcentage_total = np.round((nombre_population /population_totale_2017) * 100, 2)

# Affichage du nombre théorique de personnes pouvant être nourries et le pourcentage correspondant de la population mondiale
print(f"Il serait théoriquement possible de nourrir environ {nombre_population} personnes, ce qui représenterait environ {pourcentage_total}% de la population mondiale.")
```

```python
# Données pour le graphique
names = ['pop pouvant être nourrie', 'population mondiale']
values = [pourcentage_total, 100]

# Création du DataFrame
data = {'Taux': names, 'Values': values}

# Création de la figure avec Plotly Express
fig = px.bar(data, x='Taux', y='Values', color='Taux', color_discrete_sequence=px.colors.qualitative.Pastel)

# Personnalisation de la mise en page
fig.update_layout(
    template='plotly_dark',
    title='Représentation de la population pouvant être nourrie en 2017',
    xaxis_title='',
    yaxis_title='Population en Pourcentage')

# Affichage de la légende
fig.update_traces(showlegend=True)
# Afficher l'image dynamique
fig.show()
# Redimensionner l'image 
fig.update_layout(width=1240)
# Enregistrer en tant qu'image
fig.write_image("figure/2.png")
# Afficher l'image sur le depot Github
display(Image(filename="figure/2.png"))
```

## 2.3 - Le nombre théorique de personnes qui pourraient être nourries avec des produits végétales

```python
# Calcul de la disponibilité alimentaire totale en millions de kcal/jour pour les produits d'origine végétale
dispo_alim_population['Disponibilité alimentaire en million (Kcal/jour)'] = dispo_alim_population.apply(
    lambda row: np.round(row['Disponibilité alimentaire (Kcal/personne/jour)'] * row["Nombre d'habitants"] / 1e6, 2),
    axis=1
)
dispo_alim_population
```

```python
# Calcul de la disponibilité alimentaire totale des produits d'origine végétale en millions de kcal/jour
dispo_alim_vegetal = dispo_alim_population[dispo_alim_population['Origine'] == 'vegetale']['Disponibilité alimentaire en million (Kcal/jour)'].sum()

# Besoin calorique journalier en kcal
besoin_calorique_journalier = 2500

# Calcul du nombre de personnes pouvant être nourries avec des produits d'origine végétale
dispo_alim_vegetal_personne = dispo_alim_vegetal / besoin_calorique_journalier

# Conversion en milliards
dispo_alim_vegetal_personne_milliard = dispo_alim_vegetal_personne / 1e3

# Calcul de la proportion par rapport à la population mondiale
proportion_dispo_alim_vegetal = (dispo_alim_vegetal_personne_milliard * 1e9 / population_totale_2017) * 100

# Arrondi des valeurs
dispo_alim_vegetal_personne_milliard = round(dispo_alim_vegetal_personne_milliard, 3)
proportion_dispo_alim_vegetal = round(proportion_dispo_alim_vegetal, 2)

# Affichage du résultat
print(f"On peut nourrir environ {dispo_alim_vegetal_personne_milliard} milliard d'habitants avec des produits d'origine végétale, ce qui représente environ {proportion_dispo_alim_vegetal}% de la population mondiale.")
```

```python
# Données pour le graphique
categories = ['pop pouvant être nourrie avec des produits végétales', 'pop pouvant être nourrie', 'population mondiale'] 
values = [proportion_dispo_alim_vegetal, pourcentage_total, 100]

# Création du DataFrame
data = {'Categories': categories, 'Pourcentage de la population': values}
df = pd.DataFrame(data)

# Création du graphique avec Plotly Express
fig = px.bar(df, x='Categories', y='Pourcentage de la population', color='Categories',
             color_discrete_sequence=px.colors.qualitative.Pastel)

# Utilisation du template pour l'arrière-plan noir et le texte en blanc
fig.update_layout(title='Représentation de la population pouvant être nourrie en 2017',
                  xaxis_title=None,
                  yaxis_title='Population en Pourcentage',
                  template='plotly_dark',   # Arrière-plan noir et texte en blanc
                  showlegend=True)          # Légende

# Afficher l'image dynamique
fig.show()
# Redimensionner l'image 
fig.update_layout(width=1240)
# Enregistrer en tant qu'image
fig.write_image("figure/3.png")
# Afficher l'image sur le depot Github
display(Image(filename="figure/3.png"))
```

## 2.4 - Utilisation de la disponibilité intérieure 

- **Disponibilité intérieure** = **Production** + **Importations** + **Variation** - **Exportations** = **Nourriture** + **Aliments animaux** + **Semences** + **Pertes** + **Traitement** + **Autres utilisations**.

### a - Calcul de la part attribué à l'alimentation animales

```python
# Calcul de la somme de la disponibilité intérieure
dispo_interieure_total = donnees_disponibilite_alimentaire['Disponibilité intérieure'].sum()

# Calcul de la somme de l'alimentation pour animaux
dispo_alim_animale = donnees_disponibilite_alimentaire['Aliments pour animaux'].sum()

# Calcul de la proportion d'utilisation de la nourriture pour les animaux
proportion_animale = np.round((dispo_alim_animale / dispo_interieure_total) * 100, 2)

# Affichage du résultat
print(f"{proportion_animale}% de la disponibilité intérieure utilisée est pour l'alimentation des animaux.")
```

### b - Calcul de la part attribué à l'alimentation humaines

# 6 - Conclusions et Solutions

**Conclusions**

- La production mondiale pourait sans problème nourir toute la population mondiale.
- Les ressources sont mal répartis à travers le monde.
- L'afrique est le continent le plus touché suivi de l'Asie et l'Amérique du sud.
- L'Europe se place trés bien face au reste du monde ainsi que l'Amérique du Nord et L'Océanie
- Les pays en zone de conflits ainsi que les pays qui subissent beaucoup de catastrophes naturelles sont les plus impactés par la sous-nutrition.
- Cette étude nous permet de prendre conscience de l'étendu de la sous-nutrition au niveau mondial, elle permet aussi de pouvoir réagir et cibler les pays les plus démunis dans un avenir qui ne s'arrange pas d'un point de vue démographique et nutritionnel, climatique et politico-sociale.

**Solutions**

- Continuer à aider les pays en situation d'insécurité alimentaire (stabilisation de leur agriculture, création d'un cadre légal de coopération entre un organisme neutre comme l' ONU et les pays concernés)
- Trouver des culture alternatives et adaptables aux changement climatiques.
- Permettre aux pays ayant une grande partie de sa population en insécurité alimentaire d'être plus compétitifs face aux pays riches et cela ne se fera qu'avec la prise de conscience de la part de la population et on encourageant et favorisant le commerce équitable.
- Mettre en place des plans de préventions contre la mal nutrition de toutes ses formes.
